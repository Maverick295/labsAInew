```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000,
                           n_features=2,
                           n_informative=2,
                           n_redundant=0,
                           n_classes=2,
                           class_sep=1,
                           random_state=1)
```

```python
X.shape, y.shape
```

```python
pd.DataFrame(X).head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
y_pred_lr = model.predict(X)
```

```python
model.score(X, y)
```

```python
xx, yy = np.meshgrid(
    np.arange(X.min(axis=0)[0] - 1, X.max(axis=0)[0] + 1, 0.01),
    np.arange(X.min(axis=0)[1] - 1, X.max(axis=0)[1] + 1, 0.01))
```

```python
XX = np.array(list(zip(xx.ravel(), yy.ravel()))).reshape((-1, 2))
```

```python
Z = model.predict_proba(XX)[:,0].reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.8, levels=20)
plt.scatter(X[y==0, 0], X[y==0, 1], marker="o", c='r', s=20)
plt.scatter(X[y==1, 0], X[y==1, 1], marker="x", c='b', s=20)
```
